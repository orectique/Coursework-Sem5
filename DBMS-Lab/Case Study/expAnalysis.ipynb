```python
import pandas as pd
import json
```

```python
match = pd.read_csv('./Data Files/match.csv')
players = pd.read_csv('./Data Files/player.csv')
```

```python
match
```

```python
netPlayers = list(match['Player_White']) + list(match['Player_Black'])
```

```python
playerCounts = {}
```

```python
for player in netPlayers:
    if player in playerCounts:
        playerCounts[player] += 1
    else:
        playerCounts[player] = 1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(playerCounts.values(), bins=range(min(playerCounts.values()), max(playerCounts.values()) + 1, 1))
```
